```python
"""
Download and store IASI L3 products
"""

from wekeo_iasi_l3.download import get_IASI_products

start_day = "2024-08-15"
end_day = "2024-08-16"

files = get_IASI_products(
    start_date=start_day,
    end_date=end_day,
)
```

## Explore IASI Data Structure

First, let's explore what variables are available in the IASI .nat file:

```python
from wekeo_iasi_l3.reader_l2.reader import explore_product, IASI_L2_VARIABLES

# Show common variables that might be available
print("Common IASI L2 variables:")
for var, desc in IASI_L2_VARIABLES.items():
    print(f"  {var:30s} - {desc}")

explore_product(files[0])
```

## Read IASI Data into xarray Dataset

Now let's read specific variables from the IASI file:

```python
from wekeo_iasi_l3.reader_l2.reader import read_iasi_l2

# Read the first downloaded file with default variables
if files:
    first_file = files[0]
    print(f"Reading: {first_file}")
    
    # Read with default variables
    ds = read_iasi_l2(first_file, variables=["INTEGRATED_CO"])
    
    print(f"\nDataset dimensions: {ds.dims}")
    print(f"\nData variables:")
    for var in ds.data_vars:
        print(f"  {var}: {ds[var].shape}")
    
    print(f"\nCoordinates:")
    for coord in ds.coords:
        print(f"  {coord}: {ds[coord].shape}")
    
    # Display the dataset
    ds
```
